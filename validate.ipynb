```python
import os
import random
random.seed(a=8)
from keras.models import load_model
from dataprocess.validation import Validation
from dataprocess.table_process import load_data
from dataprocess.source import RealSource
PATH = os.path.join(os.path.expanduser('~') , 'Desktop', 'energydisagg')
```

```python
HOUSES = [2,5,39]
CHANNELS = ['main','fridge', 'bottle warmer','air conditioner','television','washing machine']

data_to_memory, house_prob, activation_prob = load_data(HOUSES, os.path.join(PATH, 'data', 'III'))

real_source = RealSource(data_to_memory = data_to_memory, channels = CHANNELS, seq_length=60, 
                        houses = HOUSES, houses_prob  = house_prob, activations_prob = activation_prob, num_seq_per_batch = 160)

main, targets = real_source._get_batch()
while main is None or targets is None:
    main, targets = real_source._get_batch()
```

```python
for i in ['III_20180717_10_59_dae_8_fridge_washing machine.h5']:
    try:
        print(i)
        model = load_model(os.path.join(PATH, 'models', i))
        print(model.summary())
        validate = Validation(main, targets, model, ['main','fridge','washing machine'], False)
        #validate._zeros_guess()
        validate._model_guess()
        #validate._plot(os.path.join(PATH,'fig','v'))
    except:
        pass
```
